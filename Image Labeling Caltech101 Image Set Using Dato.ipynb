### Image Similarity Through Magnitude of Feature Differences

> Intuitively, images that look similar to each other should have smaller differences across their extracted features than images that are as similar. Let us verify this below by subtracting features of images from each other and plotting the absolute differences.

> So these photos of faces should have smaller differences than photos with faces and a chair

```python
images['image'][16:17].show()
images['image'][21:22].show()
images['image'][11:12].show()
```

```python
print "Feature differences between 2 faces"
figure(figsize=(22,4))
da = list(images[16]['extracted_features'])
db = list(images[21]['extracted_features'])
diff = [abs(vb-va) for vb,va in zip(db,da)]
plot(range(4096), diff, marker='+', color='b')
xlim((0, 4096))
ylim((0,12))
```

```python
print "Feature differences between a face and a chair"
figure(figsize=(22,4))
da = list(images[16]['extracted_features'])
db = list(images[11]['extracted_features'])
diff = [abs(vb-va) for vb,va in zip(db,da)]
plot(range(4096), diff, marker='+', color='b')
xlim((0, 4096))
ylim((0,12))
```

### Finding Similarity Using KNN

```python
nearest_neighbor_model = graphlab.nearest_neighbors.create(images, features=['extracted_features'])
```

```python
similar_images = nearest_neighbor_model.query(images, k = 2)
```

```python
similar_images
```

> We do some cleaning to remove the instances where the query equals the reference. This happened beacause the query set was identical to the reference set.

```python
similar_images = similar_images[similar_images['query_label'] != similar_images['reference_label']]
```
